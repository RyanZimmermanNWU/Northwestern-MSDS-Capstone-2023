# Import Dataset

```python
pd.set_option('display.max_columns', None)

import pandas as pd

directory = r'https://raw.githubusercontent.com/mattlibonati/Northwestern-MSDS-Capstone-2023/main/Data/'

data = pd.read_csv(directory + 'final_data_segmented.csv')

print(data.shape)
data.head()
```

# Plot Data

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 3,figsize=(10,4))

axs[0].hist(data.median_age)
axs[0].set_title('median_age')
axs[1].hist(data.median_home_value)
axs[1].set_title('median_home_value')
axs[2].hist(data.emp_location_quotient_financial_activities)
axs[2].set_title('emp_location_quotient_financial_activities')

fig.tight_layout()

# Values by Cluster
pd.set_option('display.max_columns', None)
data.groupby('final_log_mm_k')[['median_age','median_home_value','emp_location_quotient_financial_activities']].describe()
```

# "Custom Search" - Young Finance Professional

```python
min_median_age, max_median_age = input('Min Median Age: '), input('Max Median Age: ') 
min_median_home_value, max_median_home_value = input('Min Median Home Value: '), input('Max Median Home Value: ')
min_emp_quotient_fin = input('Min Finance Employment Quotient: ')

recommendation = data[(data['median_age'] >= float(min_median_age)) & (data['median_age'] <= float(max_median_age)) \
                     & (data['median_home_value'] >= float(min_median_home_value)) & (data['median_home_value'] <= float(max_median_home_value)) \
                     & (data['emp_location_quotient_financial_activities'] >= float(min_emp_quotient_fin))]
display(recommendation.head())

# Create Recommendation
grp = recommendation.groupby('final_log_mm_k')[['final_log_mm_k']].count()
grp.rename(columns = {'final_log_mm_k' : 'counts'} ,inplace = True)
grp = grp.reset_index().sort_values(by='counts',ascending = False)
cluster_recommendation = grp.head(1)[['final_log_mm_k']].values[0][0]
print(f'Cluster Recommendation: {cluster_recommendation}')
display(data[data['final_log_mm_k'] == cluster_recommendation]) 
```
